```python
cap =cv2.VideoCapture(0) # here 0 means by deaflut webcam
while True:
    ret,frame=cap.read() # cap.read will return two values onr is ret and otheris frame ret is boolen avlue and ret can be false if there is no webam or web cam has not started and ret value is false 
    # we can aslo used to convert video to gray scale
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if ret==False:
        continue
    cv2.imshow("videoframe",frame) # 
    cv2.imshow("grayframe",gray_frame)
    # wait for user input u will press q abd you will stop the loop
    key_pressed=cv2.waitKey(1) & 0xFF # here cv.waitkey will give a return value that which key is preessed
    # but the it return 32 bit value but we want to convert into 8 bits that why we have use single and with 0xFF as 
    # 0xFF is 8 ones so it will give 8 bits value
    if key_pressed == ord('q'): # if 8 bits value matches with ascii of q then loop will become false and it will break
        break
cap.release() # it will realese all windows
cv2.destroyAllWindows() # it will destroy all window
```

```python
import cv2
import numpy as np
import os
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(train_data, queryPoint, k=5):
    vals = []
    m=train_data.shape[0]

    for i in range(m):
        X=train_data[i,:-1]
        Y=train_data[i,-1]
        d = distance(queryPoint, X)
        vals.append([d,Y])

    vals = sorted(vals)

    vals = vals[:k]

    vals = np.array(vals)

    new_vals = np.unique(vals[:,1], return_counts=True)
    print(new_vals)

    index = new_vals[1].argmax()
    pred = new_vals[0][index]

    return pred
# INIT CAMERA
cap=cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
skip=0
face_data=[] # x values for dataset
dataset_path='./final dataset/'
label=[] # yvalues for data
class_id=0 # labels for a given file
name={} #mapping  btw id - name

# data prepartion
for fx in os.listdir(dataset_path):# to get the all the files which is in data folder
    if fx.endswith('.npy'):
        #mapping btw id and name
        name[class_id]=fx[:-4]
        data_item=np.load(dataset_path+fx)
        face_data.append(data_item)
        
        # create labels for the class
        target=class_id*np.ones((data_item.shape[0]))
        class_id+=1
        label.append(target)
face_dataset=np.concatenate(face_data,axis=0)
face_label=np.concatenate(label,axis=0).reshape((-1,1))
print(face_dataset)
print(face_label)
trainset=np.concatenate((face_dataset,face_label),axis=1)
print(trainset)

# testing

while True:
    ret,frame=cap.read()
    if ret == False:
        continue
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(frame,1.3,5)
    for face in faces[-1:]:
        x,y,w,h=face
        offset=10
        face_section = frame[y-offset:y+h+offset,x-offset:x+w+offset]
        face_section = cv2.resize(face_section,(100,100)) 
        #predictted
        out=knn(trainset,face_section.flatten(),61)
        # disply on screen the name and rectengle
        pred_name=name[int(out)]
        cv2.putText(frame,pred_name,(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2,cv2.LINE_AA  )
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)
    cv2.imshow("faces",frame)
    key_pressed=cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
cap.release() # it will realese all windows
cv2.destroyAllWindows() # it will destroy all window
    

```
